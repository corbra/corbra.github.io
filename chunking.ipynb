```python
import logging
import os
#from abc import ABC

class Device():

    def __init__(self,name:str,parametermap:str):
        logging.basicConfig()
        self.log = logging.getLogger()
        #self.log.setLevel(logging.DEBUG)

        self.parametermap = parametermap
        self.name = name

    def _remove_unmapped(self,device_params:pd.DataFrame):
        # try this device_params[device_params.name.isin(parameter_defaults.keys())]
        #device_params[~device_params.name.str.contains('InternetGatewayDevice.')]
        return device_params[device_params.name.isin(parameter_defaults.keys())]

    def add_missing_parameters(self,device_params:pd.DataFrame):

        parameter_defaults_s = pd.Series(parameter_defaults.keys())

        missing_list = parameter_defaults_s[~parameter_defaults_s.isin(device_params.name)]

        for missing in missing_list:
            value = parameter_defaults.get(missing)
            new_param = pd.DataFrame({'created':[device_params.created.iloc[0]],'name':[missing],'productclass':[device_params.productclass.iloc[0]],'serial':[device_params.serial.iloc[0]],'value':[value]})
            device_params = pd.concat([device_params, new_param], ignore_index = True, axis = 0)

        return device_params

        #pd.merge(device_params,device_params.created, how="left", on=["name"])
        #parameter_defaults

    def make_mappings(self,device_params:pd.DataFrame):
        #self.log.info(f'Mapping {device_params.created.iloc[0]},{device_params.productclass.iloc[0]},{device_params.serial.iloc[0]}')

        device_params = device_params.replace(self.parametermap, regex=True)

        #device_params = self._remove_unmapped(device_params)

        return device_params


    def process_sum_active(self,device_params:pd.DataFrame,param_search_string:str)->pd.DataFrame:

        '''
        The active matches items are Summed. 
        '''
        # Find the number of connected LAN devices
        df_active  = device_params.name.str.contains(param_search_string,regex=True)
        value = (device_params[df_active].value == '1').sum()

        # create new parameter with count of devices
        new_param = pd.DataFrame({'created':[device_params.created.iloc[0]],'name':[self.parametermap[param_search_string]],'productclass':[device_params.productclass.iloc[0]],'serial':[device_params.serial.iloc[0]],'value':[value]})

        # remove all existing parameters
        device_params = device_params[~device_params.name.str.contains(param_search_string)]

        #add the new one
        device_params  = pd.concat([device_params,new_param])

        return device_params

    # def process_sum(self,device_params:pd.DataFrame,param_search_string:str)->pd.DataFrame:

    #     '''
    #     The matched items values are Summed. 
    #     '''
    #     # Find the number of connected LAN devices
    #     df_active  = device_params.name.str.contains(param_search_string,regex=True)
    #     value = device_params[df_active].value.astype(int).sum()

    #     # create new parameter with count of devices
    #     new_param = pd.DataFrame({'created':[device_params.created.iloc[0]],'name':[self.parametermap[param_search_string]],'productclass':[device_params.productclass.iloc[0]],'serial':[device_params.serial.iloc[0]],'value':[value]})

    #     # remove all existing parameters
    #     device_params = device_params[~device_params.name.str.contains(param_search_string)]

    #     #add the new one
    #     device_params  = pd.concat([device_params,new_param])

    #     return device_params


    def process_aggregate(self,device_params:pd.DataFrame,param_search_string:str,type:str='SUM')->pd.DataFrame:

        '''Perform an agregation on the row data
            Finds all parameters that match the param_search_string regex and agregates.  
            How to agregate is specified by "type"
            A new record with the agregate value is added using the device mapping 
        '''
        # Find the number of connected LAN devices
        df_active  = device_params.name.str.contains(param_search_string,regex=True)

        if type == 'SUM':
            value = device_params[df_active].value.astype(int).sum()
        elif type == 'MAX':
            value = device_params[df_active].value.astype(int).max()
        else:
            return device_params

        # create new parameter with count of devices
        new_param = pd.DataFrame({'created':[device_params.created.iloc[0]],'name':[self.parametermap[param_search_string]],'productclass':[device_params.productclass.iloc[0]],'serial':[device_params.serial.iloc[0]],'value':[value]})

        # remove all existing parameters
        device_params = device_params[~device_params.name.str.contains(param_search_string)]

        #add the new one
        device_params  = pd.concat([device_params,new_param])

        return device_params

        
        

# class Device_Y687_060(Device):

#     def make_mappings(self,device_params:pd.DataFrame):
#         self.log.info(f'Mapping {device_params.created.iloc[0]},{device_params.productclass.iloc[0]},{device_params.serial.iloc[0]}')

#         #device_params = self.process_host_active(device_params)
#         device_params = self.process_sum_active(device_params,param_search_string='InternetGatewayDevice.LANDevice.*.Hosts.Host.*.Active')
#         device_params = self.process_sum_active(device_params,param_search_string='InternetGatewayDevice.WANDevice.*.WANCommonInterfaceConfig.NumberOfActiveConnections')
        
#         return super().make_mappings(device_params)

# class Device_H89T(Device):


#     def make_mappings(self,device_params:pd.DataFrame):
#         #self.log.info(f'Mapping {device_params.created.iloc[0]},{device_params.productclass.iloc[0]},{device_params.serial.iloc[0]}')

#         device_params = self.process_aggregate(device_params,param_search_string='InternetGatewayDevice.LANDevice.*.WLANConfiguration.[0-9]+.TotalBytesReceived',type='MAX')
#         device_params = self.process_aggregate(device_params,param_search_string='InternetGatewayDevice.LANDevice.*.WLANConfiguration.[0-9]+.TotalBytesSent',type='MAX')
#         return super().make_mappings(device_params)



if __name__ == '__main__': 

    data = {'created':['2022-01-19 13:00:28','2022-01-19 13:00:28','2022-01-19 13:00:28'],
        'name':['InternetGatewayDevice.WANDevice.1.Timestamp_TGT_WANNetConfigInfo.RSRQ','InternetGatewayDevice.LANDevice.*.WLANConfiguration.1.TotalBytesReceived','InternetGatewayDevice.LANDevice.*.WLANConfiguration.2.TotalBytesReceived'],
        'productclass':['H89T','H89T','H89T'],
        'serial':['test1','H89T139297956250717','test3'],
        'value':['-12dBm','300','54']}

    
    df = pd.DataFrame(data)

    data_Y687_060 = {'created':['2022-01-19 13:00:28','2022-01-19 13:00:28'],
        'name':['InternetGatewayDevice.LANDevice.1.Hosts.Host.*.Active','InternetGatewayDevice.LANDevice.1.Hosts.Host.*.Active'],
        'productclass':['Y687_060','Y687_060'],'serial':['test1','test2'],
        'value':['1','1']}
    df_Y687_060 = pd.DataFrame(data_Y687_060)
    
    device_H89T = Device('H89T',map_H89T)

    device_Y687_060 = Device('Y687_060',map_YXXX_XXX )
    
    device_params = device_Y687_060.process_sum_active(df_Y687_060,param_search_string='InternetGatewayDevice.LANDevice.*.Hosts.Host.*.Active')
    device_params = device_Y687_060.process_sum_active(df_Y687_060,param_search_string='InternetGatewayDevice.WANDevice.*.WANCommonInterfaceConfig.NumberOfActiveConnections')
    df_Y687_060 = device_Y687_060.make_mappings(df_Y687_060)
    #print(device_Y687_060.add_missing_parameters(df_Y687_060).to_string())

    device_params = device_H89T.process_aggregate(df,param_search_string='InternetGatewayDevice.LANDevice.*.WLANConfiguration.[0-9]+.TotalBytesReceived',type='MAX')
    device_params = device_H89T.process_aggregate(df,param_search_string='InternetGatewayDevice.LANDevice.*.WLANConfiguration.[0-9]+.TotalBytesSent',type='MAX')
    df = device_H89T.make_mappings(df)
    #print(device_H89T.add_missing_parameters(df).to_string())
```

```python
import numpy as np
from datetime import datetime
import time

class Devices:

    def __init__(self,parameter_data_file:str = 'DeviceParameters/Jan26_Data.txt',speed_test_data_file:str = 'SpeedTestData/2022-01-27_09-01-29-637.csv',outfile:str='device_parameter_data.csv', chunk_size:int = 100,num_chunks:int = 1,device_class:list = []):

        logging.basicConfig()
        self.log = logging.getLogger()
        self.log.setLevel(logging.INFO)

        self.parameter_data_file = parameter_data_file
        self.outfile = outfile
        self.chunk_size = chunk_size
        self.num_chunks = num_chunks
        self.device_class = device_class
        self.speed_test_data_file = speed_test_data_file
        self.start = True

    def _remove_outfile(self):

        #remove output file
        if(os.path.exists(self.outfile) and os.path.isfile(self.outfile)):
            os.remove(self.outfile)
            self.log.info(f"{self.outfile} file deleted")
        else:
            self.log.info(f"{self.outfile} file not found")   

    def _process_parameters(self,device_params:pd.DataFrame):
        key = (device_params.created.iloc[0],{device_params.productclass.iloc[0]},{device_params.serial.iloc[0]})
        
        self.log.debug(f'processing {device_params.created.iloc[0]},{device_params.productclass.iloc[0]},{device_params.serial.iloc[0]}')
       

        if device_params.productclass.iloc[0] in ['Y687-060','Y464-067','Y687-060_e7','Y687-060_HT']:
            device = Device(device_params.productclass.iloc[0],map_YXXX_XXX ) 

                    #device_params = self.process_host_active(device_params)
            device_params = device.process_sum_active(device_params,param_search_string='InternetGatewayDevice.LANDevice.*.Hosts.Host.*.Active')
            device_params = device.process_sum_active(device_params,param_search_string='InternetGatewayDevice.WANDevice.*.WANCommonInterfaceConfig.NumberOfActiveConnections')

            device_params = device.make_mappings(device_params)
            device_params = device._remove_unmapped(device_params)
            try:
                device_params = device.add_missing_parameters(device_params)
            except IndexError as e:
                self.log.error(f'{e}, key{key}:')
            
        if device_params.productclass.iloc[0] in ['H89T','H89T_e7','HVR-879T','HVR-879TG','H89T_e7','HVR-879T_e7','AOG_K74','AOG_H87_KOFH']:
            device= Device(device_params.productclass.iloc[0],map_H89T)
            device_params = device.process_aggregate(device_params,param_search_string='InternetGatewayDevice.LANDevice.*.WLANConfiguration.[0-9]+.TotalBytesReceived',type='MAX')
            device_params = device.process_aggregate(device_params,param_search_string='InternetGatewayDevice.LANDevice.*.WLANConfiguration.[0-9]+.TotalBytesSent',type='MAX')
            device_params = device.make_mappings(device_params)
            device_params = device._remove_unmapped(device_params)
            try:
                device_params = device.add_missing_parameters(device_params)
            except IndexError as e:
                self.log.error(f'{e}, key{key}:')

        # if device_params.productclass.iloc[0] == 'Y464-067': 
        #     device = Device('Y464-067',map_y_type)
        #     device_params = device.process_sum_active(device_params,param_search_string='InternetGatewayDevice.LANDevice.*.Hosts.Host.*.Active')
        #     device_params = device.process_sum_active(device_params,param_search_string='InternetGatewayDevice.WANDevice.*.WANCommonInterfaceConfig.NumberOfActiveConnections')
        #     device_params = device.make_mappings(device_params)
        #     device_params = device._remove_unmapped(device_params)
        #     try:
        #         device_params = device.add_missing_parameters(device_params)
        #     except IndexError as e:
        #         self.log.error(f'{e}, key{key}:')

        try:
            #Ignore for no mappings
            if device_params.productclass.iloc[0] in NO_MAPPINGS:
                print(f'Device {device_params.productclass.iloc[0]} has no mappings')
                #break
        except IndexError as e:
                self.log.error(f'{e}, key{key}')
                self.log.debug(f'{device_params}')
        
        return device_params

    def write_to_file(self,devices:pd.DataFrame,speed_test_data:pd.DataFrame)->pd.DataFrame:
        # do ÃŸpivot
        try:
            devices = devices.drop_duplicates(subset=['created', 'productclass','serial','name'])
            devices = devices.pivot(index=['created', 'productclass','serial'],columns=['name'],values=['value'])
            devices = devices.reset_index()
        except (ValueError,KeyError) as e:
            self.log.error(f'{e}')
            self.log.debug(f'{devices.to_string()}')

        #flatten pivot - messy, review!
        columns = devices.columns.get_level_values(1).to_list()
        columns[0] = 'created'
        columns[1] = 'productclass'
        columns[2] = 'serial'
        devices.columns = columns
        
        # Merge to add Target download speed
        devices_speed_test_data = pd.merge_asof(devices.sort_values('created'),speed_test_data.sort_values('Time Stamp'),
            left_on=['created'],right_on = ['Time Stamp'],
            left_by = ['productclass','serial'],right_by = ['Product Class','Serial'],
            direction='forward',tolerance=pd.Timedelta('1h'))
        
        # Remove records with no download speed
        devices_speed_test_data = devices_speed_test_data.dropna(subset=['Serial_y', 'Product Class', 'Download Speed (Mbps)','Time Stamp'])

        if self.start:
            devices_speed_test_data.to_csv(self.outfile, mode='a', index=False, header=True)
            self.start = False
        else:
            devices_speed_test_data.to_csv(self.outfile, mode='a', index=False, header=False)
        
        devices = pd.DataFrame()

        return devices

    def process_file(self):
        device_params = []
        devices = pd.DataFrame()
        self._remove_outfile()

        count = self.num_chunks

        speed_test_data = pd.read_csv(self.speed_test_data_file,
            na_values = ['-'],
            dtype={'Serial': str,'Product Class':str,'Download Speed (Mbps)':float,'Time Stamp':str},on_bad_lines='skip'  )

        speed_test_data = speed_test_data.dropna() 
        speed_test_data['Time Stamp'] = pd.to_datetime(speed_test_data['Time Stamp'])

        self.log.info(f'\nspeed test file {self.speed_test_data_file} stats\ncount = {speed_test_data.shape[0]}\nmin speed test date: {speed_test_data["Time Stamp"].min()}\nmax speed test date: {speed_test_data["Time Stamp"].max()}')

        for chunk in tqdm(pd.read_csv(self.parameter_data_file, delimiter='\t',chunksize=self.chunk_size,parse_dates=['created'])):

            # Get device list
            distinct = chunk[['created', 'productclass', 'serial']].drop_duplicates().reset_index()

            #self.log.info(f'\nprocessing chunk of {len(chunk)} parameters\ndates between: {chunk.created.min()}\nmax speed test date: {chunk.created.max()}', end = "\r")

            # loop device list
            for index, row in distinct.iterrows():

                # get the parmaeters that belong to this device
                bmask = chunk.isin({'created':[row.created],'productclass':[row.productclass],'serial':[row.serial]})

                bmask = bmask.created & bmask.productclass & bmask.serial

                # check for devices split accross chunks
                if len(device_params) and (index == 0)  and \
                        (device_params.created.iloc[0] == chunk[bmask].created.iloc[0])  \
                        and (device_params.productclass.iloc[0] == chunk[bmask].productclass.iloc[0])  \
                        and (device_params.serial.iloc[0] == chunk[bmask].serial.iloc[0]):

                        device_params = pd.concat([chunk[bmask],device_params])
                else:
                    device_params = chunk[bmask]

                if index == len(distinct)-1: #  skip to next chunck check for more data for this device
                    continue

                #check filter
                if device_params.productclass.iloc[0] not in self.device_class:
                    continue

                device_params = self._process_parameters(device_params)

                devices = pd.concat([devices,device_params])

            if count != -1: # if its -1, loop till finished
                count -= 1

            if count ==0:
                break
        
            # skip if there is nothing to write
            if len(devices) == 0:
                continue
            else:
                devices = self.write_to_file(devices,speed_test_data)


if __name__ == '__main__': 
    start_time = time.time()

    # 'Y687-060_e7','Y687-060_HT','Y464-067'
    # 'H89T_e7','HVR-879T','HVR-879TG','H89T_e7','HVR-879T_e7','AOG_K74','AOG_H87_KOFH'
    include_models = ['Y687-060_HT','H89T','Y687-060','Y464-067','Y464-067','Y687-060_e7','H89T_e7','HVR-879T','HVR-879TG','H89T_e7','HVR-879T_e7','AOG_K74','AOG_H87_KOFH']

    devices = Devices(parameter_data_file= 'DeviceParameters/Feb02_Data.txt.zip',
        speed_test_data_file = 'SpeedTestData/2022-01-27_09-01-29-637.csv',
        num_chunks = 50,device_class=include_models,chunk_size=2000,outfile='DEVSP-2022-01-26.csv')

    # devices = Devices(parameter_data_file= 'DeviceParameters/Mar03_Data.txt.zip',
    #     speed_test_data_file = 'SpeedTestData/2022-01-13_16-44-53-776.csv',
    #     num_chunks = 20,device_class=['H89T','Y687-060','Y464-067'],chunk_size=2000,outfile='DEVSP-2022-01-26.csv')

    # devices = Devices(parameter_data_file= 'DeviceParameters/Mar03_Data.txt.zip',
    #     speed_test_data_file = 'SpeedTestData/2022-02-25_00-00-21-942.csv',
    #     num_chunks = 20,device_class=['H89T','Y687-060','Y464-067'],chunk_size=2000,outfile='DEVSP-2022-01-26.csv')

    devices.process_file()
    print("--- %s seconds ---" % (time.time() - start_time))
    #../SpeedTestData/2022-01-26_09-20-10-199.csv

    # Matched files
    # 2022-01-19: DeviceParameters/Jan26_Data.txt.zip SpeedTestData/2022-01-20_09-01-22-282.csv
    # 2022-01-26: DeviceParameters/Feb02_Data.txt.zip SpeedTestData/2022-01-27_09-01-29-637.csv
    # 2022-01-12: DeviceParameters/Jan19_Data.txt.zip SpeedTestData/2022-01-13_16-44-53-776.csv
    # 2022-01-05: DeviceParameters/Jan12_Data.txt.zip SpeedTestData/2022-01-06_08-28-35-309.csv Too Small!!
    # 2022-02-24: DeviceParameters/Mar03_Data.txt.zip SpeedTestData/2022-02-25_00-00-21-942.csv Bad Data skipped
```
